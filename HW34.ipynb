```python
import pandas as pd
import plotly
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import sweetviz as sv
import ydata_profiling as yp

# dataframe
db = pd.read_excel('cities.xls')
num = [i+1 for i in range(20)]
db.index = num
print(db, '\n', type(db))
```

```python
# 2 graphs in 1 figure

fig = make_subplots(rows=1, cols=2)
fig1 = go.Scatter(x=db['площа'], y=db['населення'], 
                  mode='markers', marker=dict(size=10, color="Red"), name='населення від площі міста')
fig2 = go.Scatter(x=db['населення'], y=db['середн_дохід'], 
                  mode='markers', marker=dict(size=10, color="LightSeaGreen"), name='середн_дохід від населення')

fig.add_trace(fig1, row=1, col=1)
fig.add_trace(fig2, row=1, col=2)
fig.update_traces(marker_line_width=2)  #, marker_size=10)
fig.update_layout(title_text="Рисунок 1А", title_font_size=30)
fig.show(renderer="png")
# fig.write_html("Fig1A.html")
```

```python
fig = make_subplots()
fig.add_trace(go.Scatter(y=db['середн_дохід'], x=db['населення'], mode='markers', name='середн_дохід, грн'))
fig.add_trace(go.Scatter(y=db['площа'], x=db['населення'], mode='markers', name='площа, кв.км'))

fig.update_traces(marker_line_width=2, marker_size=10)
fig.update_layout(title='Рисунок 1Б', title_font_size=30, legend=dict(x=0.05, y=0.95, font_size=18))
fig.update_xaxes(title='Населення', title_font_size=18)
fig.show(renderer="png")
# fig.write_html("Fig1B.html")
```

```python
# corr matrix and heatmap

db1 = pd.DataFrame(db.loc[:, ['населення', 'середн_дохід', 'площа']])
corr = db1.corr()
print(corr)

fig = px.imshow(corr, text_auto=True)
fig.show(renderer="png")
# fig.write_html("HeatmapA.html")

fig = go.Figure()
fig.add_trace(go.Heatmap(x=corr.columns, y=corr.index, z=corr, colorscale='Blues'))
fig.show(renderer="png")
# fig.write_html("HeatmapB.html")
```
